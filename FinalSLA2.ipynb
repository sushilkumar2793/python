```python
# Add a new column "Communication Status" with NaN values and ensure it's of type object
SLAReport['Communication Status'] = pd.Series(dtype='object')

# Define the custom function to determine the Communication Status
def determine_communication_status(row):
    if row['Status'] == 'Success':
        return 'Success'
    elif row['Status'] == 'Inactive':
        return 'Inactive'
    elif row['Status'] == 'Failure':
        if row['Last Update Date'] == '-':
            return 'Never'
        else:
            return 'Non'
    else:
        return 'Unknown'  # In case there are other statuses not covered

# Apply the custom function to each row to create the new column
SLAReport['Communication Status'] = SLAReport.apply(determine_communication_status, axis=1)


# Save the updated DataFrame back to the Excel file
SLAReport.to_excel('/content/gdrive/MyDrive/DATA SCIENCE/SLAReport090824final.xlsx', sheet_name='Sheet1')
SLAReport['Communication Status'].value_counts(dropna=False)
```

```python
# prompt: now i want to graph

import matplotlib.pyplot as plt

# Group the data by Communication Status and count the occurrences
status_counts = SLAReport['Communication Status'].value_counts()

# Extract labels and values from the Series
labels = status_counts.index.to_list()
values = status_counts.to_list()

# Create a bar chart
fig, ax = plt.subplots()
ax.bar(labels, values)

# Customize the chart
plt.xlabel("Communication Status")
plt.ylabel("Count")
plt.title("Communication Status Distribution")
plt.show()
```

```python
# prompt: i want sim wise communication graph

import matplotlib.pyplot as plt
# Group the data by Sim Type and Communication Status
grouped_data = SLAReport.groupby(['Sim Type', 'Communication Status']).size().unstack()

# Create a bar chart to visualize the grouped data
grouped_data.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Sim Type')
plt.ylabel('Count')
plt.title('Communication Status Distribution by Sim Type')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the bar chart
plt.show()
```

```python
import pandas as pd

# Read the SLAReport090824final.xlsx file
SLAReport = pd.read_excel('/content/gdrive/MyDrive/DATA SCIENCE/SLAReport090824final.xlsx')

# Replace 'Three phase' with 'Three Phase'
SLAReport['Meter Type'] = SLAReport['Meter Type'].replace('Three phase', 'Three Phase')

# Count meter types
meter_type_counts = SLAReport['Meter Type'].value_counts().reset_index()
meter_type_counts.columns = ['Meter Type', 'Count']

# Save the updated data and counts to the same Excel sheet
with pd.ExcelWriter('/content/gdrive/MyDrive/DATA SCIENCE/SLAReport090824final.xlsx') as writer:
    SLAReport.to_excel(writer, sheet_name='Sheet1', index=False)  # Save original data without index
    meter_type_counts.to_excel(writer, sheet_name='Meter Type Counts', index=False)  # Save counts on a new sheet

# Print counts for clarity (optional)
print("Meter Type Counts:")
for index, row in meter_type_counts.iterrows():
    print(f"{row['Meter Type']}: {row['Count']}")



```
